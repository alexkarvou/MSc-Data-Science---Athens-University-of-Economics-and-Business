At first we need to generate some data to work with. To make things more interesting and for the (), we used a multivariate normal distribution for dimension 100, with mean 0 vector and 

```r
library(MASS)
mu<-rep(0,200)
Sigma<-matrix(.2,nrow=200,ncol=200)+diag(200)*0.15
dataset<-as.data.frame(mvrnorm(n=100000,mu = mu,Sigma=Sigma))
dataset$psi<-0
coeffs<-rnorm(200,4,2.7)
for (i in 1:dim(dataset)[1]){
    dataset$psi[i]<-drop(coeffs %*% as.matrix(dataset)[i,1:200])+rnorm(1,0,8)
}
```
