# sklearn

```python
#X=my_data.drop('Time_Deposits_Flag',axis=1)
X=my_data[['Internet_Trans_Num','Insurance_Balances']]
Y=my_data.Time_Deposits_Flag
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=13)
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn import tree
Log_reg=LR()
Log_reg.fit(X_train,y_train)
y_pred_log=Log_reg.predict(X_test)
y_score_log=Log_reg.predict_proba(X_test)
tree=tree.DecisionTreeClassifier()
tree = tree.fit(X_train, y_train)
y_pred_tree=tree.predict(X_test)
y_score_tree=tree.predict_proba(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
%matplotlib inline

print (classification_report(y_test, y_pred_log))
print ("The accuracy score is {:.2%}".format(accuracy_score(y_test, y_pred_log)))
print (classification_report(y_test, y_pred_tree))
print ("The accuracy score is {:.2%}".format(accuracy_score(y_test, y_pred_tree)))   
# Compute ROC curve and area under the curve
fpr, tpr, thresholds = roc_curve(y_test, y_score_log[:, 1])
roc_auc = auc(fpr, tpr)
print ("Area under the ROC curve : %f" % roc_auc)

# Plot ROC curve
plt.clf()
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic example')
plt.legend(loc="lower right")
```

```python
import pandas as pd
my_data=pd.read_csv("mark_data.csv")
my_data.head()
```

```python
my_data.Time_Deposits_Flag=my_data.Time_Deposits_Flag.replace('F','0')
my_data.Time_Deposits_Flag=my_data.Time_Deposits_Flag.replace('T','1')
my_data.Time_Deposits_Flag=pd.to_numeric(my_data.Time_Deposits_Flag)
Y=my_data.Time_Deposits_Flag
my_data=my_data.drop(my_data.columns[0],axis=1)
my_data=my_data.drop('Time_Deposits_Flag',axis=1)
X=my_data
from sklearn.preprocessing import MinMaxScaler
X = MinMaxScaler().fit_transform(X)
#X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2,random_state=3 )
```

```python
from collections import Counter
from imblearn.under_sampling import RandomUnderSampler
os=RandomUnderSampler(random_state=90)
X_str,y_str=os.fit_sample(X_train,y_train)
print('Resampled dataset shape {}'.format(Counter(y_str)))
```

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
Log_reg=LR(C=1.2)
Log_reg.fit(X_str,y_str)
y_pred_log=Log_reg.predict(X_test)
y_score_log=Log_reg.predict_proba(X_test)
Tree=tree.DecisionTreeClassifier(max_depth=4,min_samples_leaf=3)
Tree = Tree.fit(X_str, y_str)
y_pred_tree=Tree.predict(X_test)
y_score_tree=Tree.predict_proba(X_test)
rf=RandomForestClassifier(max_depth=2,min_samples_leaf=3,n_estimators=70)
rf.fit(X_str,y_str)
y_pred_rf=rf.predict(X_test)
y_score_rf=rf.predict_proba(X_test)
ada=AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=4,min_samples_leaf=3),algorithm='SAMME',n_estimators=70)
ada.fit(X_str,y_str)
y_pred_ada=ada.predict(X_test)
y_score_ada=ada.predict_proba(X_test)
gb=GradientBoostingClassifier(max_depth=10,n_estimators=100,random_state=10,min_samples_split=100,
                             min_samples_leaf=30,learning_rate=.01,max_features=3)
gb.fit(X_str,y_str)
y_pred_gb=gb.predict(X_test)
y_score_gb=gb.predict_proba(X_test)
nn=MLPClassifier(hidden_layer_sizes=(500,30),activation='tanh',solver='sgd')
nn.fit(X_str,y_str)
y_pred_nn=nn.predict(X_test)
y_score_nn=nn.predict_proba(X_test)
```

```python
gb=GradientBoostingClassifier(max_depth=10,n_estimators=100,random_state=10,min_samples_split=100,
                             min_samples_leaf=30,learning_rate=.01,max_features=3)
gb.fit(X_str,y_str)
y_pred_gb=gb.predict(X_test)
y_score_gb=gb.predict_proba(X_test)
y_gb_1=gb.predict(X_train)
y_gb_2=gb.predict(X_test)
print('Gradient boosting for train')
print (classification_report(y_train, y_gb_1))
print('Gradient boosting for test')
print (classification_report(y_test, y_gb_2))
```
