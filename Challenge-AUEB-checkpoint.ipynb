```python
import networkx as nx
import numpy as np
import os
import zipfile
import csv
from scipy.sparse import lil_matrix
G = nx.read_weighted_edgelist('edgelist.csv', delimiter='\t', create_using=nx.DiGraph())


categories = os.listdir('train_webpages')

# Create a dictionary that maps webpages to the categories they belong
train_labels = {}
for category in categories:
    webpages = os.listdir('train_webpages/'+category)
    for webpage in webpages:
        train_labels[webpage] = category

# Create a numpy array containing the categories of the training examples
y_train=[]
for k in train_labels.values():
    y_train.append(k)
#cd C:/Users/karvo/Desktop/aueb/data Science Challenge/Proj_data
# Create a dictionary that maps each category to an integer
label2index = {}
y_unique = np.unique(y_train)
for i in range(y_unique.size):
    label2index[y_unique[i]] = len(label2index)

# Create the training matrix. Its rows correspond to the webpages of the training set
# and its columns to the 12 categories. Given a webpage, each column is set equal to the
# number of neighbors (from the graph) of the webpage that belong to the category mapped 
# to that column		
X_train = np.zeros((len(train_labels), 3*len(label2index)))
i = 0
for webpage in train_labels:
    aux0=[]
    for neighbor in G.neighbors(webpage):
        if neighbor in train_labels and neighbor!=webpage:#children
            X_train[i, label2index[train_labels[neighbor]]] += 1
        #spouses
        #aux0.extend(G.predecessors(neighbor))
    #for j in aux0:
    #     if j in train_labels and j !=webpage:
    #        X_train[i, label2index[train_labels[j]]+36] += 1     
    aux1=G.predecessors(webpage)
    for j in aux1:#parents
        if j in train_labels and j !=webpage:
            X_train[i, label2index[train_labels[j]]+12] += 1
    aux2=[]
    for j in aux1:#siblings
        aux2.extend(G.neighbors(j))
    for j in aux2:
        if j in train_labels and j!=webpage:
            X_train[i, label2index[train_labels[j]]+24] += 1 
    #if webpage!=neighbor:
                #X_train[i, label2index[train_labels[neighbor]]+12] += G.edge[webpage][neighbor]['weight']
    #if X_train[i,12:24].sum()!=0:         
        #X_train[i,12:24]=X_train[i,12:24]/X_train[i,12:24].sum()              
            
    i += 1

# Read webpages of the test set
test_webpages = os.listdir('test_webpages')

# Create the test matrix in the same way as in the case of the training matrix
X_test = np.zeros((len(test_webpages), 3*len(label2index)))
for i in range(len(test_webpages)):
    for neighbor in G.neighbors(test_webpages[i]):
        if neighbor in train_labels and neighbor!=test_webpages[i]:
            X_test[i, label2index[train_labels[neighbor]]] += 1
   #     aux0.extend(G.predecessors(neighbor))
    #for j in aux0:
     #    if j in train_labels and j !=test_webpages[i]:
      #      X_test[i, label2index[train_labels[j]]+36] += 1
    aux1=G.predecessors(test_webpages[i])
    for j in aux1:
        if j in train_labels and j !=test_webpages[i]:
            X_test[i, label2index[train_labels[j]]+12] += 1
    aux2=[]
    for j in aux1:
        aux2.extend(G.neighbors(j))
    for j in aux2:
        if j in train_labels and j!=test_webpages[i]:
            X_test[i, label2index[train_labels[j]]+24] += 1         
            #if test_webpages[i]!=neighbor:
                #X_test[i, label2index[train_labels[neighbor]]+12] +=G.edge[test_webpages[i]][neighbor]['weight']
   #if X_test[i,12:24].sum()!=0:        
        #X_test[i,12:24]=X_test[i,12:24]/X_test[i,12:24].sum()        
print ("Train matrix dimensionality: ", X_train.shape)
print ("Test matrix dimensionality: ", X_test.shape)
```

```python
import numpy as np# MPA
#grandparents
X_aux=np.zeros(X_train.shape[0]*24).reshape(X_train.shape[0],24)
Xtest_aux=np.zeros(X_test.shape[0]*24).reshape(X_test.shape[0],24)
for webpage in train_labels:
    for child in G.neighbors(webpage):
        if child!=webpage:
            for granchild in G.neighbors(child):
                if granchild in train_labels and granchild!=child:
                    X_aux[i, label2index[train_labels[granchild]]] += 1
    for parent in G.predecessors(webpage):
        if parent !=webpage:
            for granparent in G.predecessors(parent):
                if granparent in train_labels and granparent!=parent:
                    X_aux[i,label2index[train_labels[granparent]]+12]+=1

    
for i in range(len(test_webpages)):
    for child in G.neighbors(test_webpages[i]):
        if child in train_labels and child!=test_webpages[i]:
            for granchild in G.neighbors(child):
                if granchild in train_labels and granchild!=child:
                    Xtest_aux[i, label2index[train_labels[granchild]]] += 1
                    
    for parent in G.predecessors(test_webpages[i]):
        if parent!=test_webpages[i]:
            for granparent in G.predecessors(parent):
                if granparent in train_labels and granparent !=parent:
                    Xtest_aux[i, label2index[train_labels[granparent]]+12] += 1

X_train=np.hstack((X_train,X_aux))
X_test=np.hstack((X_test,Xtest_aux))
```

```python
def splits(iterable):
    s=tuple(iterable)
    for pos in range(2,len(s)):
        if len(iterable[0:pos])>2:
            yield iterable[0:pos]
        if len(iterable[pos:])>2:    
            yield iterable[pos:]
```
